```python
from simulator import HathorSimulator, Miner
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
```

```python
sim = HathorSimulator(block_weight=10)

m0 = Miner(hashpower=10000000)
sim.add_miner(m0)
```

```python
sim.run(3600*24*30)
print('block weight', sim.block_weight)
```

```python
sim.block_weight
```

```python
prev = None
v = []
for x in sim.blocks[4000:20000]:
    if prev is None:
        prev = x
    else:
        v.append(x.time - prev.time)
        prev = x
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.hist(v, bins='fd', density=True);

x = np.linspace(0, max(v), 1000)
y = scipy.stats.expon.pdf(x, 0, 128)
plt.plot(x, y, 'r', linewidth=2);

plt.ylabel('Histogram')
plt.xlabel('Time between blocks with one miner');

plt.xlim(0, 800);
```
