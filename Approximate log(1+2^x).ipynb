```python
from sympy import symbols, diff, log, factorial
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
```

```python
y = log(1 + 2**(-x))/log(2)
```

```python
dy = [y]
for i in range(15):
    dy.append(diff(dy[-1], x))
```

```python
dy
```

```python
dy0 = [f.subs({x: 2.5}) for f in dy]
```

```python
dy0
```

```python
coef = [a/factorial(i) for i, a in enumerate(dy0)]
```

```python
coef
```

```python
def calc(coef, x):
    r = 0
    for a in reversed(coef):
        r = r*x + float(a)
    return r
```

```python
x_values = np.linspace(0, 40, 1000)
y_values = [math.log(1+2**(-a))/math.log(2) for a in x_values]
#y2_values = [calc(coef[:5], a-2.5) for a in x_values]
```

```python
plt.figure(figsize=(8, 6), dpi=300)
plt.plot(x_values, y_values, label='original')
#plt.plot(x_values, y2_values, alpha=0.5, label='approx')
plt.legend();
plt.ylim(0, 0.001);
plt.xlim(10, max(x_values));
```
